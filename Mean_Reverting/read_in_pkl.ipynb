```python
import pickle

with open('log_rt_backtest_99clcoint.pkl', 'rb') as f:
    log_rts = pickle.load(f)
    
for i in log_rts:
    print(i, len(log_rts[i]), sum(log_rts[i])/len(log_rts[i]))
    # print median
    print(i, len(log_rts[i]), sorted(log_rts[i])[len(log_rts[i])//2])
```

```python
# print porportion of positive return

for i in log_rts:
    print(i, len([j for j in log_rts[i] if j > 0])/len(log_rts[i]))
```

```python
# max and min of log_rt

for i in log_rts:
    print(i, max(log_rts[i]), min(log_rts[i]))
    
```

```python
# plot the distribution of log_rt in each threshold

import matplotlib.pyplot as plt

for i in log_rts:
    plt.hist(log_rts[i], bins=100)
    plt.title('threshold = ' + str(i))
    plt.show()
    
```
